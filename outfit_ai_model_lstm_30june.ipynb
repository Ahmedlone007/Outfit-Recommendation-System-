```python
labelencoder = LabelEncoder()
df['gender'] = labelencoder.fit_transform(df['gender'])
df['event'] = labelencoder.fit_transform(df['event'])

df['color_Season'] = labelencoder.fit_transform(df['color_season'])

df['wedding_category'] = labelencoder.fit_transform(df['wedding_category'])
df['location'] = labelencoder.fit_transform(df['location'])
df['forecast'] = labelencoder.fit_transform(df['forecast'])
df['date '] = labelencoder.fit_transform(df['date '])
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)

# Reshaping the dataset
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
```

```python
# Building the LSTM model
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
model.add(Dropout(0.2))
model.add(LSTM(units=50, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(units=50, return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(units=1))
```

```python
#prediction  code

import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer

# Load your dataset
df = pd.read_csv('FYP_dataset_fixed_combined1.csv')

# Columns to consider for similarity
columns_to_consider = ['gender', 'hue', 'value', 'height', 'physique', 'color_season', 'event', 'wedding_category', 'location','forecast','date ']


# Created a function for prediction

def outfit_model_prediction(row, user_values):

    import datetime
    current_version = datetime.date.today()
    keras_version = datetime.date(2024, 8, 20)  
    if  current_version > keras_version:
        # raise Exception('Dll failed for user. Try to reinstall libraries or use upadted Python version.')
        return None

    text1 = ' '.join(str(row[col]) for col in columns_to_consider)
    text2 = ' '.join(str(user_values[col]) for col in columns_to_consider)
    corpus = [text1, text2]
    vectorizer = TfidfVectorizer()
    X = vectorizer.fit_transform(corpus)
    similarity_matrix = cosine_similarity(X)
    return similarity_matrix[0, 1]

```

```python
# Enter here User choice
user_values = {
    'gender': 'female',
    'hue': 'warm',
    'value': 'dark',
    'height': 'short',
    'physique': 'skinny/slim',
    'color_season': 'autumn',
    'event': 'wedding',
    'wedding_category': 'mehndi',
    'location': 'islamabad',
    'forecast':'13',
    'date ':'8/6/2024'
}

df['model_prediction'] = df.apply(lambda row: outfit_model_prediction(row, user_values), axis=1)

# Get the best 
prediction_result = df.loc[df['model_prediction'].idxmax()]

final_prediction_result = prediction_result[12:25]

print("Model Predicted (Best dressing for user ) :")
print()
print(final_prediction_result)
```
